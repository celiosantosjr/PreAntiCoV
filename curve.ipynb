```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(1, 2, figsize=(14, 5), dpi=600)
ensemble = 'EasyEnsemble'
is_fs = {'2020-06-18_13-08': 'Yes', '2020-06-18_16-07': 'No'}
for name in ['2020-06-18_13-08', '2020-06-18_16-07']:
    prcurve = np.load("./results_classification/{:s}/Pre-Cls/RF/Default/pr.npy".format(name), 
                      allow_pickle=True).item()
    precision = prcurve["precision"]
    recall = prcurve["recall"]
    aucpr = prcurve['aucpr']
    ax[0].plot(recall, precision, lw=1.2, label="{:s} (AUC = {:.2f} %)".format(is_fs[name], aucpr * 100))
    ax[0].set_title("Precision-Recall Curve")
    ax[0].set_xlabel("recall")
    ax[0].set_ylabel("precision")
    rocurve = np.load("./results_classification/{:s}/Pre-Cls/RF/Default/roc.npy".format(name), 
                      allow_pickle=True).item()
    fpr = rocurve['fpr']
    tpr = rocurve['tpr']
    aucroc = rocurve['aucroc']
    ax[1].plot(fpr, tpr, lw=1.2, label="{:s} (AUC = {:.2f} %)".format(is_fs[name], aucroc * 100))
    ax[1].set_title("Receiver Operating Curve")
    ax[1].set_xlabel("fpr")
    ax[1].set_ylabel("tpr")
    
ax[0].legend(title='Rank-sum')
ax[1].plot([0, 1], [0, 1], color='#808080', linestyle='-.', label='Random')
ax[1].legend(title='Rank-sum')
fig.savefig("prroc.png")
```
